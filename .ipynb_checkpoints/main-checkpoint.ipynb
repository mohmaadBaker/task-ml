```python
import pandas as pd
```

```python
data=pd.read_csv('imdb_top_1000.csv')
data
```

```python
print(data.columns)
```

```python
# col=['Meta_score']
# data.drop(columns=col, inplace=True)
```

```python
genre_counts = data['Genre'].value_counts()

# Print the genre counts
print(genre_counts)
```

```python
genre_counts = data['Genre'].value_counts()

# Create a new dataframe from the genre counts
genre_counts_df = pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})
print(genre_counts_df)
```

```python
import matplotlib.pyplot as plt
genre_counts = data['Genre'].value_counts()
genre_counts_df = pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})
genre_counts_df = genre_counts_df.sort_values(by='Count', ascending=False)
genre_counts_df.plot(x='Genre', y='Count', kind='bar', legend=False)

# Set the labels and title
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Count of Movies/TV Shows in Each Genre')

# Show the plot
plt.show()
```

```python
director_counts = data['Director'].value_counts()

# Create a new dataframe from the director counts
director_counts_df = pd.DataFrame({'Director': director_counts.index, 'Count': director_counts.values})
director_counts_df = director_counts_df.sort_values(by='Count', ascending=False)
```

```python
import matplotlib.pyplot as plt
director_counts = data['Director'].value_counts()

director_counts_df = pd.DataFrame({'Director': director_counts.index, 'Count': director_counts.values})
director_counts_df = director_counts_df.sort_values(by='Count', ascending=False)
director_counts_df.plot(x='Director', y='Count', kind='bar', legend=False)

plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Count of Movies/TV Shows Directed by Each Director')

# Show the plot
plt.show()
```
